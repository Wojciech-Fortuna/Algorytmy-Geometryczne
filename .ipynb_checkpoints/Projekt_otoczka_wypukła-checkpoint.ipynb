# Narzędzia do punktów

Tworzenie własnego zbioru punktów przy pomocy pygame. Zamknięcie programu przyciskiem "x" spowoduje zwrócenie pustej tablicy. Punkty są zapisywane w kolejności ich podawania.

```python
def Position_changer(points):
    new_points = []
    for i in range (len(points)):
        new_point = (points[i][0], -points[i][1]+550)
        new_points.append(new_point)
    return new_points


def Set_sections_using_mouse():    
    pygame.init()
    
    SCREEN_WIDTH=800
    SCREEN_HEIGHT=600
    
    screen=pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
    
    points=[]
    run=True
    
    while run:
        buttom=pygame.Rect((0,560,120,40))
        pygame.draw.rect(screen, (255,0,0), buttom)
        font = pygame.font.SysFont(None, 24)
        img = font.render('End', True, (255,255,255))
        screen.blit(img, (40, 570))
        empty_space=pygame.Rect((120,560,680,40))
        pygame.draw.rect(screen, (255,255,255), empty_space)
        
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                points.clear()
                run=False
            if event.type == pygame.MOUSEBUTTONUP:
                pos = pygame.mouse.get_pos()
                if 0<=pos[0]<=120 and 560<pos[1]<=600:
                    run=False
                    time.sleep(0.5)
                    break
                elif pos[1]<=560:
                    pygame.draw.circle(screen,(255,0,0),pos,5)
                    points.append(pos)
    
        pygame.display.update()
    
    pygame.quit()
    
    points = Position_changer(points)
    return points
```

```python
vis.clear()
vis.add_point(Set_sections_using_mouse())
vis.show()
```

Zapisywanie punktów w pliku csv

```python
def save_points_to_csv(points, filename):
    with open(filename, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerows(points)
```

Wczytywanie punktów z pliku csv

```python
def read_points_from_csv(filename):
    points = []
    with open(filename, 'r') as csvfile:
        csv_reader = csv.reader(csvfile)
        for row in csv_reader:
            point = tuple(map(float, row))
            points.append(point)
    return points
```

```python
def save_points_to_txt(points, filename):
    with open(filename, 'w') as txtfile:
        for point in points:
            txtfile.write(f"{point[0]}, {point[1]}\n")
```

# Generowanie punktów na płaszczyźnie

Generowanie unkalnych losowych punktów w przestrzeni 2D w obrębie kwadratu

```python
def generate_uniform_points(a=(0, 0), b=(100, 0), c=(100, 100), d=(0, 100), n=100):
    random.seed()
    points = set()
    for i in range(n):
        while True:
            x = random.uniform(a[0],b[0])
            y = random.uniform(a[1],d[1])
            new_point = (x,y)
            if new_point not in points:
                points.add(new_point)
                break
    return list(points)
        
    """
    Funkcja generuje równomiernie n punktów na kwadwratowym
    obszarze od left do right (jednakowo na osi y)
    o współrzędnych rzeczywistych
    :param left: lewy kraniec przedziału
    :param right: prawy kraniec przedziału
    :param n: ilość generowanych punktów
    :return: tablica punktów w postaci krotek współrzędnych
             np. [(x1, y1), (x2, y2), ... (xn, yn)]
    """
    raise Exception("Unimplemented")
```

Generowanie równomiernie rozłożonych punktów w przestrzeni 2D leżących na okręgu

```python
def generate_circle_points(O = (0,0), R = 10, n=100):
    random.seed()
    points = set()
    for i in range (n):
        while True:
            t = random.uniform(0,n) * np.pi * 2 / n
            x = R*np.cos(t) + O[0]
            y = R*np.sin(t) + O[1]
            new_point = (x,y)
            if new_point not in points:
                points.add(new_point)
                break
    return list(points)
    """
    Funkcja generuje jednostajnie n punktów na okręgu o środku O i promieniu R
    :param O: krotka współrzędnych x, y określająca środek okręgu
    :param R: promień okręgu
    :param n: ilość generowanych punktów
    :return: tablica punktów w postaci krotek współrzędnych
    """
    raise Exception("Unimplemented")
```

```python
vis.clear()
vis.add_point(generate_circle_points())
vis.show()
```

Generowanie unikalnych losowych punktów w przestrzeni 2D leżących na obwodzie prostokąta

```python
def generate_rectangle_points(a=(-10, -10), b=(10, -10), c=(10, 10), d=(-10, 10), n=100):
    random.seed()
    points = set()
    edge = c[0]-a[0]
    for i in range (n):
        while True:
            dist=random.uniform(0,edge*4)
            if dist<edge:
                x=a[0]
                y=dist+a[1]
            elif dist<edge*2:
                x=dist-edge+d[0]
                y=d[1]
            elif dist<edge*3:
                x=b[0]
                y=dist-(edge*2)+b[1]
            else:
                x=dist-(edge*3)+a[0]
                y=a[1]
            new_point = (x,y)
            if new_point not in points:
                points.add(new_point)
                break
    return list(points)
            
    
    '''
    Funkcja generuje n punktów na obwodzie prostokąta
    o wierzchołkach w punktach a, b, c i d
    :param a: lewy-dolny wierzchołek prostokąta
    :param b: prawy-dolny wierzchołek prostokąta
    :param c: prawy-górny wierzchołek prostokąta
    :param d: lewy-górny wierzchołek prostokąta
    :param n: ilość generowanych punktów
    :return: tablica punktów w postaci krotek współrzędnych
    '''
    raise Exception("Unimplemented")
```

Generowanie punktów w przestrzeni 2D w następujący sposób: 
1) po tyle samo unikalnych losowych punktów na dwóch bokach kwadratu leżących na osiach, które nie są jednocześnie wierzchołkami tego kwadratu
2) po tyle samo uikalnych losowych punktów na przekątnych kwadratu, które nie są jednocześnie wierzchołkami tego kwadratu
3) punkty stanowiące wierzczhołki tego kwadratu

```python
def generate_square_points(a=(0, 0), b=(10, 0), c=(10, 10), d=(0, 10),
                           axis_n=25, diag_n=20):
    random.seed()
    points = set()
    points.add(a)
    points.add(b)
    points.add(c)
    points.add(d)
    for i in range (axis_n):
        while True:
            x = a[0]
            y = random.uniform(a[1],d[1])
            new_point = (x,y)
            if new_point not in points:
                points.add(new_point)
                break
    for i in range (axis_n):
        while True:
            x = random.uniform(a[0],b[0])
            y = a[0]
            new_point = (x,y)
            if new_point not in points:
                points.add(new_point)
                break
    for i in range (diag_n):
        while True:
            x = random.uniform(a[0],c[0])
            y = x
            new_point = (x,y)
            if new_point not in points:
                points.add(new_point)
                break
    for i in range (diag_n):
        while True:
            x = random.uniform(d[0],b[0])
            y = 10-x
            new_point = (x,y)
            if new_point not in points:
                points.add(new_point)
                break
    return list(points)
    
    
    '''
    Funkcja generuje axis_n punktów na dwóch bokach kwadratu
    leżących na osiach x i y oraz diag_n punktów na
    przektąnych kwadratu, którego wyznaczają punkty
    a, b, c i d.
    :param a: lewy-dolny wierzchołek kwadratu
    :param b: prawy-dolny wierzchołek kwadratu
    :param c: prawy-górny wierzchołek kwadratu
    :param d: lewy-górny wierzchołek kwadratu
    :param axis_n: ilość generowanych punktów na każdym
                   z dwóch boków kwadratu równoległych do osi x i y
    :param diag_n: ilość generowanych punktów na każdej
                   przekątnej kwadratu
    :return: tablica punktów w postaci krotek współrzędnych
    '''
    raise Exception("Unimplemented")
```

Generowanie punktów w przestrzeni 2D w następujący sposób:
1) unikalne losowe punkty w obrębie kwadraratu
2) równomiernie rozłożone punkty na okręgu o środku w tym samym punkcie co środek kwadratu, którego promień w stosunku do długości boku kwadratu wynosi 1.5

```python
def even_circle_points(O, R, n):
    points=[]
    step = 1/n
    for i in range (n):
        t = step * i
        x = R*np.cos(t*np.pi*2) + O[0]
        y = R*np.sin(t*np.pi*2) + O[1]
        points.append((x,y))
    return points
```

```python
def set_convex_hull(inside = 6, outside = 4, size_of_square = 20, O=(0,0)):
    a = [-size_of_square/2+O[0], -size_of_square/2+O[1]]
    b = [size_of_square/2+O[0], -size_of_square/2+O[1]]
    c = [size_of_square/2+O[0], size_of_square/2+O[1]]
    d = [-size_of_square/2+O[0], size_of_square/2+O[1]]
    new_points=generate_uniform_points(a,b,c,d, inside)
    new_points.extend(even_circle_points(O, size_of_square*1.5, outside))
    new_points=random.sample(new_points,len(new_points))
    return new_points
```

# Algorytmy

## Algorytm Grahama

```python
def graham_algorithm(Q, epsilon = 1e-25, clockwise = False):
    
    def Start_point (Q):
        start=Q[0]
        start_idx=0
        for i in range (1,len(Q)):
            if Q[i][1]<start[1]:
                start=Q[i]
                start_idx=i
            elif Q[i][1]==start[1]:
                if Q[i][0]<start[0]:
                    start=Q[i]
                    start_idx=i
        return start,start_idx
    
    
    def orientation (a,b,c,epsilon):
        if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
            b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
            c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1
    
    
    def further (a,b,c):
        if a[0]==b[0]:
            if abs(a[1]-b[1])>abs(a[1]-c[1]):
                return True
            else:
                return False
        else:
            if abs(a[0]-b[0])>abs(a[0]-c[0]):
                return True
            else:
                return False
    
    
    def ComparePoints(start_point, first, second, epsilon):
        orient=orientation(start_point, first, second, epsilon)
        if orient>0 or (orient==0 and further(start_point, first, second)==False):
            return -1
        else:
            return 1
        
    
    def Sort_points (Q,start,start_idx,epsilon):
        if len(Q)==1: return Q
        start_point = Q[start_idx]
        Q.pop(start_idx)    
        Q = sorted(Q, key = functools.cmp_to_key(lambda first, second: ComparePoints(start_point, first, second, epsilon)))
        Q_sorted=[start_point]
        for i in range (len(Q)-1):
            if orientation(start_point,Q[i],Q[i+1],epsilon)!=0:
                Q_sorted.append(Q[i])
        Q_sorted.append(Q[len(Q)-1])
        return Q_sorted
    
    
    def Get_convex_hull(Q,epsilon):
        convex_hull=[Q[0],Q[1],Q[2]]
        i=3
        t=2
        while i<len(Q):
            if orientation(convex_hull[t-1],convex_hull[t],Q[i],epsilon)>0:
                convex_hull.append(Q[i])
                t+=1
                i+=1
            else:
                convex_hull.pop()
                t-=1
        return convex_hull


    points = copy.deepcopy(Q)
    start,start_idx=Start_point(points)
    points=Sort_points(points,start,start_idx,epsilon)
    if len(points)<3:
        convex_hull = points
    else:
        convex_hull = Get_convex_hull(points,epsilon)
    if clockwise == True:
        convex_hull.reverse()
    return convex_hull

    
    '''
    Funkcja buduje otoczkę wypukłą dla podanego
    zbioru punktów Q algorymem Grahama
    :parm Q: zbiór punktów
    :return: tablica punktów w postaci krotek współrzędnych
    '''
    raise Exception("Unimplemented")
```

```python
def graham_algorithm_draw(Q,epsilon = 1e-25):
    
    def Start_point (Q):
        start=Q[0]
        start_idx=0
        for i in range (1,len(Q)):
            if Q[i][1]<start[1]:
                start=Q[i]
                start_idx=i
            elif Q[i][1]==start[1]:
                if Q[i][0]<start[0]:
                    start=Q[i]
                    start_idx=i
        return start,start_idx
    
    
    def orientation (a,b,c,epsilon):
        if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
            b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
            c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1
    
    
    def further (a,b,c):
        if a[0]==b[0]:
            if abs(a[1]-b[1])>abs(a[1]-c[1]):
                return True
            else:
                return False
        else:
            if abs(a[0]-b[0])>abs(a[0]-c[0]):
                return True
            else:
                return False
    
    
    def ComparePoints(start_point, first, second, epsilon):
        orient=orientation(start_point, first, second, epsilon)
        if orient>0 or (orient==0 and further(start_point, first, second)==False):
            return -1
        else:
            return 1
        
    
    def Sort_points (Q,start,start_idx,epsilon):
        if len(Q)==1: return Q
        start_point = Q[start_idx]
        Q.pop(start_idx)    
        Q = sorted(Q, key = functools.cmp_to_key(lambda first, second: ComparePoints(start_point, first, second, epsilon)))
        Q_sorted=[start_point]
        Q_eliminated=[]
        for i in range (len(Q)-1):
            if orientation(start_point,Q[i],Q[i+1], epsilon)!=0:
                Q_sorted.append(Q[i])
            else:
                Q_eliminated.append(Q[i])
        Q_sorted.append(Q[len(Q)-1])
        return Q_sorted, Q_eliminated
    
    
    def Draw_sorted_points(Q,epsilon):
        start,start_idx=Start_point(Q)
        green,red=Sort_points(Q,start,start_idx,epsilon)
        vis.clear()
        if len(green)>1:
            vis.add_line_segment([(green[i], green[0]) for i in range(1, len(green))])
            vis.add_point(green[1:], color=['green'])
        vis.add_point(red, color=['red'])
        vis.add_point(green[0], color=['purple'])
        vis.show()
        return green,red
    
    
    def Less_than_4_draw(Q,not_convex_hull):
        vis.clear()
        if len(not_convex_hull)==0:
            convex_hull=Q.copy()
        else:
            convex_hull=[]
            for i in range (len(Q)):
                if not_convex_hull[0]!=Q[i]:
                    convex_hull.append(Q[i])
        if len(convex_hull)==3:
            vis.add_line_segment([(Q[0], Q[1]),(Q[1], Q[2]),(Q[2], Q[0])] , color=['blue'])
        elif len(convex_hull)==2:
            vis.add_line_segment([(Q[0], Q[1])] , color=['blue'])
        vis.add_point(convex_hull, color=['blue'])
        if len(not_convex_hull)!=0:
            vis.add_point(not_convex_hull, color=['orange'])
        vis.show()
    
    
    def Vis_hull_graham(convex_hull,not_convex_hull,Q,to_start,is_good=False,start=False,last=False):
        vis.clear()
        if start==True:
            vis.add_line_segment([(convex_hull[i], convex_hull[i+1]) for i in range(2)], color=['blue'])
            vis.add_point(convex_hull, color=['blue'])
        elif last==True:
            vis.add_line_segment([(convex_hull[i], convex_hull[i+1]) for i in range(len(convex_hull)-1)], color=['blue'])
            vis.add_line_segment([(convex_hull[len(convex_hull)-1], convex_hull[0])], color=['blue'])
            vis.add_point(convex_hull, color=['blue'])
        else:
            vis.add_line_segment([(convex_hull[i], convex_hull[i+1]) for i in range(len(convex_hull)-2)], color=['blue'])
            vis.add_point(convex_hull[:-1], color=['blue'])
            if is_good==True:
                vis.add_line_segment([(convex_hull[-2], convex_hull[-1]),(convex_hull[-1], Q[to_start])] , color=['green'])
                vis.add_point([convex_hull[-1]], color=['green'])
            else:
                vis.add_line_segment([(convex_hull[-2], convex_hull[-1]),(convex_hull[-1],  Q[to_start])] , color=['red'])
                vis.add_point([convex_hull[-1]], color=['red'])
            vis.add_point([Q[to_start]], color=['purple'])
        vis.add_point(not_convex_hull, color=['orange'])
        if to_start<len(Q)-1:
            vis.add_point(Q[to_start+1:], color=['black'])
        vis.show()
    

    def Draw_convex_hull(Q,eliminated,epsilon):
        not_convex_hull=eliminated.copy()
        if len(Q)<4:
            Less_than_4_draw(Q,not_convex_hull)
        else:
            convex_hull=[Q[0],Q[1],Q[2]]
            i=3
            t=2
            Vis_hull_graham(convex_hull,not_convex_hull,Q,i-1,start=True)
            while i<len(Q):
                if orientation(convex_hull[t-1],convex_hull[t],Q[i],epsilon)>0:
                    Vis_hull_graham(convex_hull,not_convex_hull,Q,i,is_good=True)
                    convex_hull.append(Q[i])
                    t+=1
                    i+=1
                else:
                    Vis_hull_graham(convex_hull,not_convex_hull,Q,i,is_good=False)
                    not_convex_hull.append(convex_hull[-1])
                    convex_hull.pop()
                    t-=1
            Vis_hull_graham(convex_hull,not_convex_hull,Q,i,last=True)


    points = copy.deepcopy(Q)
    vis.clear()
    vis.add_point(points, color=['black'])
    vis.show()
    green,red = Draw_sorted_points(points,epsilon)
    Draw_convex_hull(green,red,epsilon)

    '''
    Funkcja buduje otoczkę wypukłą dla podanego
    zbioru punktów Q algorymem Grahama oraz
    pokazuje poszczególne kroki algorymu
    :parm Q: zbiór punktów
    :return: krotka w postaci (tablica punktów w postaci krotek współrzędnych,
                                wizualizer pokazujący kroki algorytmu) 
    '''
```

## Algorytm przyrostowy

```python
def incremental_algorithm (Q,epsilon = 1e-25, clockwise = False):
    
    def orientation (a,b,c, epsilon):
        if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
            b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
            c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1
            
    
    def add_to_hull(new_point,hull,is_up,epsilon):
        while True:
            if len(hull) == 1:
                hull.append(new_point)
                return hull
            else:
                if is_up == True:
                    if orientation(hull[-2],new_point,hull[-1],epsilon) > 0:
                        hull.append(new_point)
                        return hull
                    else:
                        hull.pop()
                else:
                    if orientation(hull[-2],new_point,hull[-1],epsilon) < 0:
                        hull.append(new_point)
                        return hull
                    else:
                        hull.pop()
    
    
    def merge_hull(up_hull,down_hull):
        # both hulls have at least 2 elements
        hull = []
        for i in range (len(up_hull)):
            hull.append(up_hull[i])
        for i in range (len(down_hull)-2,0,-1):
            hull.append(down_hull[i])
        return hull


    if len(Q) == 1: return Q
    points = copy.deepcopy(Q)
    points.sort()
    up_hull = [points[0]]
    down_hull = [points[0]]
    for i in range (1, len(points)):
        up_hull = add_to_hull(points[i],up_hull,True, epsilon)
        down_hull = add_to_hull(points[i],down_hull,False, epsilon)
    hull = merge_hull(up_hull,down_hull)
    if clockwise == False:
        hull.reverse()
    return hull

    

    '''
    Funkcja buduje otoczkę wypukłą dla podanego
    zbioru punktów Q algorymem Jarvisa
    :parm Q: zbiór punktów
    :return: tablica punktów w postaci krotek współrzędnych
    '''
```

```python
def incremental_algorithm_draw(Q,epsilon = 1e-25):
    
    def vis_hull_incremental(hull_1, hull_2, seg, points, seg_color, points_colors):
        vis.clear()
        before_points = []
        before_seg = []
        for i in range (len(points)):
            vis.add_point([points[i][j] for j in range (len(points[i])) if points[i][j] not in before_points], color=[points_colors[i]])
            before_points.extend(points[i])
        if len(hull_1) > 1:
            vis.add_line_segment([(hull_1[j], hull_1[j+1]) for j in range(len(hull_1)-1)], color=['blue'])
        if len(hull_2) > 1:
            vis.add_line_segment([(hull_2[j], hull_2[j+1]) for j in range(len(hull_2)-1)], color=['blue'])
        for i in range (len(seg)):
            vis.add_line_segment(seg[i], color = seg_color)
        vis.show()
    
    
    def Show_hull(points, up_hull, down_hull):
        vis.clear()
        if len(up_hull)>1:
            vis.add_line_segment([(up_hull[i], up_hull[i+1]) for i in range(len(up_hull)-1)], color=['blue'])
        if len(down_hull)>1:
            vis.add_line_segment([(down_hull[i], down_hull[i+1]) for i in range(len(down_hull)-1)], color=['blue'])
        vis.add_point(up_hull, color=['blue'])
        vis.add_point(down_hull, color=['blue'])
        vis.add_point([points[i] for i in range (len(points)) if (points[i] not in up_hull and points[i] not in down_hull)], color=['black'])
        vis.show()
    
    
    def orientation (a,b,c, epsilon):
        if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
            b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
            c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1
    
    
    def add_to_hull(new_point, hull, is_up, epsilon, all_points, second_hull, to_show):
        while True:
            merged_hull = hull + second_hull
            if len(hull) == 1:
                if to_show:
                    seg =[(hull[0],new_point)]
                    points = [[new_point],[hull[0]],merged_hull,all_points]
                    seg_color = ['green']
                    points_colors = ['orange','green','blue','black']
                    vis_hull_incremental(hull, second_hull, seg, points, seg_color, points_colors)
                hull.append(new_point)
                return hull
            else:
                seg =[(hull[-2],hull[-1]),(hull[-1],new_point)]
                points = [[new_point],[hull[-2],hull[-1]],merged_hull,all_points]
                if is_up == True:
                    if orientation(hull[-2],new_point,hull[-1],epsilon) > 0:
                        if to_show:
                            seg_color = ['green']
                            points_colors = ['orange','green','blue','black']
                            vis_hull_incremental(hull, second_hull, seg, points, seg_color, points_colors)
                        hull.append(new_point)
                        return hull
                    else:
                        if to_show:
                            seg_color = ['red']
                            points_colors = ['orange','red','blue','black']
                            vis_hull_incremental(hull, second_hull, seg, points, seg_color, points_colors)
                        hull.pop()
                else:
                    if orientation(hull[-2],new_point,hull[-1],epsilon) < 0:
                        if to_show:
                            seg_color = ['green']
                            points_colors = ['orange','green','blue','black']
                            vis_hull_incremental(hull, second_hull, seg, points, seg_color, points_colors)
                        hull.append(new_point)
                        return hull
                    else:
                        if to_show:
                            seg_color = ['red']
                            points_colors = ['orange','red','blue','black']
                            vis_hull_incremental(hull, second_hull, seg, points, seg_color, points_colors)
                        hull.pop()


    def merge_hull(up_hull,down_hull):
        # both hulls have at least 2 elements
        hull = []
        for i in range (len(up_hull)):
            hull.append(up_hull[i])
        for i in range (len(down_hull)-2,0,-1):
            hull.append(down_hull[i])
        return hull


    points = copy.deepcopy(Q)
    vis.clear()
    vis.add_point(points, color=['black'])
    vis.show()
    points.sort()
    up_hull = [points[0]]
    down_hull = [points[0]]
    Show_hull(points, up_hull, down_hull)
    for i in range (1, len(points)):
        to_show = False
        if i >= 3:
            to_show = True
        up_hull = add_to_hull(points[i],up_hull,True, epsilon, points, down_hull, to_show)
        down_hull = add_to_hull(points[i],down_hull,False, epsilon, points, up_hull, to_show)
        Show_hull(points, up_hull, down_hull)
    hull = merge_hull(up_hull,down_hull)



```

## Algorytm górnej i dolnej otoczki

```python
def andrews_monotone_chain_algorithm(Q, epsilon = 1e-25, clockwise = False):
    
    def orientation(a,b,c, epsilon):
        if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
            b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
            c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1
    
    
    def Get_convex_hull(P,epsilon):
        convex_hull=[P[0]]
        i=1
        t=0
        while i<len(P):
            if t == 0:
                convex_hull.append(P[i])
                t+=1
                i+=1
            else:
                if orientation(convex_hull[t-1],convex_hull[t],P[i],epsilon)<0:
                    convex_hull.append(P[i])
                    t+=1
                    i+=1
                else:
                    convex_hull.pop()
                    t-=1
        return convex_hull


    if len(Q)==1: return Q
    points = copy.deepcopy(Q)
    points = sorted(points, key = lambda point: (point[0],point[1]))
    convex_hull_upper = Get_convex_hull(points,epsilon)
    points.reverse()
    convex_hull_lower = Get_convex_hull(points,epsilon)
    convex_hull = convex_hull_upper
    convex_hull.pop()
    convex_hull.extend(convex_hull_lower)
    convex_hull.pop()
    if clockwise == False:
        convex_hull.reverse()
    return convex_hull


    '''
    Funkcja buduje otoczkę wypukłą dla podanego
    zbioru punktów Q algorymem Jarvisa
    :parm Q: zbiór punktów
    :return: tablica punktów w postaci krotek współrzędnych
    '''
```

```python
def andrews_monotone_chain_algorithm_draw(Q, epsilon = 1e-25):
    
    def Vis_hull_andrews(convex_hull,not_convex_hull,Q,to_start,is_good=False,start=False,last=False):
        vis.clear()
        if start==True:
            vis.add_line_segment([(convex_hull[0], Q[to_start])], color=['blue'])
            vis.add_point(convex_hull, color=['blue'])
            vis.add_point([Q[to_start]], color=['blue'])
        elif last==True:
            vis.add_line_segment([(convex_hull[i], convex_hull[i+1]) for i in range(len(convex_hull)-1)], color=['blue'])
            vis.add_point(convex_hull, color=['blue'])
        else:
            vis.add_line_segment([(convex_hull[i], convex_hull[i+1]) for i in range(len(convex_hull)-2)], color=['blue'])
            vis.add_point(convex_hull[:-1], color=['blue'])
            if is_good==True:
                vis.add_line_segment([(convex_hull[-2], convex_hull[-1]),(convex_hull[-1], Q[to_start])] , color=['green'])
                vis.add_point([convex_hull[-1]], color=['green'])
            else:
                vis.add_line_segment([(convex_hull[-2], convex_hull[-1]),(convex_hull[-1],  Q[to_start])] , color=['red'])
                vis.add_point([convex_hull[-1]], color=['red'])
            vis.add_point([Q[to_start]], color=['purple'])
        vis.add_point(not_convex_hull, color=['orange'])
        if to_start<len(Q)-1:
            vis.add_point(Q[to_start+1:], color=['black'])
        vis.show()
    
    
    def Show_hull(points, convex_hull, connecting_line = False):
        vis.clear()
        vis.add_line_segment([(convex_hull[i], convex_hull[i+1]) for i in range(len(convex_hull)-1)], color=['blue'])
        if connecting_line:
            vis.add_line_segment([(convex_hull[len(convex_hull)-1], convex_hull[0])], color=['blue'])
        vis.add_point(convex_hull, color=['blue'])
        vis.add_point([points[i] for i in range (len(points)) if points[i] not in convex_hull], color=['black'])
        vis.show()
    
    
    def orientation(a,b,c, epsilon):
        if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
            b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
            c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1
    
    
    def Get_convex_hull(P,epsilon):
        not_convex_hull = []
        convex_hull=[P[0]]
        i=1
        t=0
        while i<len(P):
            if t == 0:
                Vis_hull_andrews(convex_hull,not_convex_hull,P,i,start=True)
                convex_hull.append(P[i])
                t+=1
                i+=1
            else:
                if orientation(convex_hull[t-1],convex_hull[t],P[i],epsilon)<0:
                    Vis_hull_andrews(convex_hull,not_convex_hull,P,i,is_good=True)
                    convex_hull.append(P[i])
                    t+=1
                    i+=1
                else:
                    Vis_hull_andrews(convex_hull,not_convex_hull,P,i,is_good=False)
                    not_convex_hull.append(convex_hull[-1])
                    convex_hull.pop()
                    t-=1
        Vis_hull_andrews(convex_hull,not_convex_hull,P,i,last=True)
        return convex_hull


    points = copy.deepcopy(Q)
    vis.clear()
    vis.add_point(points, color=['black'])
    vis.show()
    points = sorted(points, key = lambda point: (point[0],point[1]))
    convex_hull_upper = Get_convex_hull(points,epsilon)
    Show_hull(points, convex_hull_upper)
    points.reverse()
    convex_hull_lower = Get_convex_hull(points,epsilon)
    Show_hull(points, convex_hull_lower)
    convex_hull = convex_hull_upper
    convex_hull.pop()
    convex_hull.extend(convex_hull_lower)
    convex_hull.pop()
    Show_hull(points, convex_hull, True)   
```

```python
def quickhull_algorithm (pkt):    

    def orient(a,b,c, epsilon=1e-25):
        #if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            #a=(decimal.Decimal('%.32f' % a[0]),decimal.Decimal('%.32f' % a[1]))
            #b=(decimal.Decimal('%.32f' % b[0]),decimal.Decimal('%.32f' % b[1]))
            #c=(decimal.Decimal('%.32f' % c[0]),decimal.Decimal('%.32f' % c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1

    
    #def orient(A, B, C, eps):
    #        det = (C[1]-B[1])*(B[0]-A[0])-(C[0]-B[0])*(B[1]-A[1])
    #        if(abs(det) <= eps):
    #            return 0
    #        if(det < 0.0):
    #            return -1
    #        return 1
        
    
    def norm(A):
        return A[0] * A[0] + A[1] * A[1]
    
    
    
    def Mergesort(l, r, points, eps, temp):
        if(l >= r):
            return
        mid = (l+r)//2
        Mergesort(l, mid, points, eps, temp)
        Mergesort(mid+1, r, points, eps, temp)
        i, j, k = l, mid+1, 0
    
        while(i <= mid and j <= r):
            o = orient((0, 0), points[i], points[j], eps)
            if(o == 1.0):
                temp[k] = points[i]
                i += 1
            elif(o == -1.0):
                temp[k] = points[j]
                j += 1
            else:
                if(norm(points[i]) > norm(points[j])):
                    temp[k] = points[j]
                    j += 1
                else:
                    temp[k] = points[i]
                    i += 1
            k += 1
    
        while(i <= mid):
            temp[k] = points[i]
            i, k = i+1, k+1
    
        while(j <= r):
            temp[k] = points[j]
            j, k = j+1, k+1
    
        for x in range(k):
            points[l+x] = temp[x]
        return
    
    
    
    def det_3x3(a, b, c, epsilon = 1e-25):
        #if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            #a=(decimal.Decimal('%.32f' % a[0]),decimal.Decimal('%.32f' % a[1]))
            #b=(decimal.Decimal('%.32f' % b[0]),decimal.Decimal('%.32f' % b[1]))
            #c=(decimal.Decimal('%.32f' % c[0]),decimal.Decimal('%.32f' % c[1]))
        return (c[1]-b[1])*(b[0]-a[0])-(c[0]-b[0])*(b[1]-a[1])
    
    
    
    
    def divide(points, L, R):
        n = len(points)
        top, bottom = [], []
        vec = (R[0] - L[0], R[1] - L[1])
        for i in range(n):
            act = points[i]
            det = det_3x3(L, R, act)
            if(det < 0):
                bottom.append(act)
            elif(det > 0):
                top.append(act)
        return bottom, top
    
    
    
    def find_high(points, L, R):
        n = len(points)
        max_point, max_pole , ind = (-1 , -1), -float('inf'), -1
        for i in range(n):
            pole = det_3x3(L, R, (points[i][0], points[i][1]))
            if(pole > max_pole):
                max_pole = pole
                max_point, ind = (points[i][0], points[i][1]), points[i][2]
        return max_point, ind
    
    
    
    def scal(tab1, tab2):
        for i in range(len(tab2)):
            tab1.append(tab2[i])
        return tab1
    
    
    def find_corners(points):
        mini_point, mini, ind_mini, maxi_point, maxi, ind_maxi = (-1, -1), float('inf'), -1, (-1, -1), -float('inf'), -1
        n = len(points)
        for i in range(n):
            x, y, jd = points[i]
            if(x > maxi or (x == maxi and y > maxi_point[1])):
                maxi_point = (x, y)
                maxi, ind_maxi = x, points[i][2]
            if(x < mini or (x == mini and y < mini_point[1])):
                mini_point = (x, y)
                mini, ind_mini = x, points[i][2]
        return mini_point, ind_mini, maxi_point, ind_maxi
    
    
    
    
    def QuickHull_rek(points, L, R, vis):
        n = len(points)
        if(n == 0):
            return
        bottom, top = divide(points, L, R) #bottom odrzucamy
        if(len(top) == 0):
            return
        H, ind = find_high(top, L, R)
        vis[ind] = 1
        QuickHull_rek(top, L, H, vis)
        QuickHull_rek(top, H, R, vis)
        
    
    
    
    def solve(points_test):
        n, points = len(points_test), []
        for i in range(n):
            points.append((pkt[i][0], pkt[i][1], i))
        vis, res, res_points = [0] * n, [], []
        L, ind1, R, ind2 = find_corners(points)
        vis[ind1], vis[ind2] = 1, 1
        QuickHull_rek(points, L, R, vis)
        QuickHull_rek(points, R, L, vis)
        for i in range(n):
            if(vis[i]):
                res.append(i)
                res_points.append((points[i][0], points[i][1]))
        return res, res_points
    
    
    def find_convex_points(hull):
        m = len(hull)
        temp = [[-1, -1] for _ in range(m)]
        found, y_min, x_min = -10, float('inf'), float('inf')
        for i in range(m):
            x, y = hull[i]
            y_min = min(y_min, y)
        for i in range(m):
            x, y = hull[i]
            if(y == y_min and x < x_min):
                x_min, found = x, i
        new_hull = []
        for i in range(m):
            new_hull.append((hull[i][0] - x_min, hull[i][1] - y_min))
        Mergesort(0, m-1, new_hull, 0, temp)
        dp = []
        for i in range(m):
            dp.append((new_hull[i][0] + x_min, new_hull[i][1] + y_min))
        return dp


    def Start_point (Q):
        start=Q[0]
        start_idx=0
        for i in range (1,len(Q)):
            if Q[i][1]<start[1]:
                start=Q[i]
                start_idx=i
            elif Q[i][1]==start[1]:
                if Q[i][0]<start[0]:
                    start=Q[i]
                    start_idx=i
        return start_idx

    
    def classify(hull):
        n = len(hull)
        A, B = hull[0], hull[-1]
        up, down = [A], []
        for i in range(1, n-1):
            orient = det_3x3(A, B, hull[i])
            if(orient > 0):
                up.append(hull[i])
            elif(orient < 0):
                down.append(hull[i])
        down.append(B)
        down.reverse()
        new_hull = up + down
        return new_hull


    if len(pkt)<3:
        #vis.clear()
        #vis.add_point(pkt, color=['red'])
        #vis.show()
        return pkt
    for i in range (len(pkt)):
        pkt[i]=(decimal.Decimal('%.32f' % pkt[i][0]),decimal.Decimal('%.32f' % pkt[i][1]))
    indexes, hull = solve(pkt)
    hull = find_convex_points(hull)
    if len(hull) == 2:
        #vis.clear()
        #vis.add_point(pkt, color=['red'])
        #vis.show()
        return hull
    n = len(hull)
    hull.sort()
    hull = classify(hull)
    idx_of_wrong = [False]*n
    for i in range (n):
        if orient(hull[i],hull[(i+1)%n],hull[(i+2)%n], epsilon=1e-25)>=0:
            idx_of_wrong[i] = True
    the_true = []
    for i in range (n):
        if idx_of_wrong[i] == False:
            the_true.append(hull[i])
    hull = the_true
    hull = find_convex_points(hull)
    for i in range (len(hull)):
        first = hull[i][0].quantize(decimal.Decimal('1e-20'), rounding=decimal.ROUND_HALF_EVEN)
        second = hull[i][1].quantize(decimal.Decimal('1e-20'), rounding=decimal.ROUND_HALF_EVEN)
        hull[i] = (first,second)
    for i in range (len(hull)):
        first = float(hull[i][0])
        second = float(hull[i][1])
        hull[i] = (first,second)

    
    #vis.clear()
    #vis.add_point(pkt, color=['blue'])
    #if len(hull)>1:
    #    vis.add_line_segment([(hull[i], hull[i+1]) for i in range(len(hull)-1)], color=['red'])
    #    vis.add_line_segment([(hull[0], hull[-1])], color=['red'])
    #if len(hull)>0:
    #    vis.add_point(hull, color=['red'])
    #vis.show()
    return hull






```

```python
def divide_and_conquer_algorithm(Q, epsilon = 1e-25, clockwise = False):
    
    def orientation(a,b,c, epsilon):
        if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
            b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
            c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1
    
    
    def merge(A, B, a, b):
        n, m, hull = len(A), len(B), []
        for i in range(a+1):
            hull.append(A[i])
        for j in range(b, m, 1):
            hull.append(B[j])
        return hull
    
    
    def concat_arr(arr1, arr2):
        res = []
        for i in range(len(arr1)):
            res.append(arr1[i])
        m = len(arr2)
        start, end = 0, m-1
        if(arr2[0] == arr1[0]):
            start = 1
        if(arr2[-1] == arr2[-1]):
            end = m-2
        for j in range(end, start-1, -1):
            res.append(arr2[j])
        return res
    
    
    def split(hull):
        n, ind = len(hull), -1
        for i in range(1, n, 1):
            if(hull[i-1] > hull[i]):
                ind = i-1
                break
        if(ind == -1):
            return hull, [hull[0], hull[-1]]
        l, r = hull[:(ind+1)], hull[(ind):]
        r.append(hull[0])
        r.reverse()
        return l, r
        
    
    def check(A, B, mid_a, mid_b, p, epsilon):
        n, m = len(A), len(B)
        if (mid_a >= 1):
            orient = orientation(A[mid_a], B[mid_b], A[mid_a - 1], epsilon)
            if (orient * p >= 0):
                return False
        if (mid_a <= n-2):
            orient = orientation(A[mid_a], B[mid_b], A[mid_a + 1], epsilon)
            if (orient * p > 0):
                return False
        if (mid_b >= 1):
            orient = orientation(A[mid_a], B[mid_b], B[mid_b - 1], epsilon)
            if (orient * p > 0):
                return False
        if (mid_b <= m-2):
            orient = orientation(A[mid_a], B[mid_b], B[mid_b + 1], epsilon)
            if (orient * p >= 0):
                return False
        return True
        
    
    def bin_search(A, B, p, epsilon):
        n, m = len(A), len(B)
        left_a, right_a, left_b, right_b, mid_a, mid_b = 0, n-1, 0, m-1, -1, -1
        mid_a = (left_a + right_a)//2
        while(True):
            left_b, right_b = 0, m-1
            while(True):
                mid_b, flag = (left_b + right_b)//2, True
                if(mid_b >= 1):
                    orient = orientation(A[mid_a], B[mid_b], B[mid_b-1], epsilon)
                    if(orient * p > 0):
                        right_b, flag = mid_b-1, False
                if(mid_b <= m-2):
                    orient = orientation(A[mid_a], B[mid_b], B[mid_b+1], epsilon)
                    if(orient * p >= 0):
                        left_b, flag = mid_b+1, False
                if(flag):
                    break
            left_a, right_a = 0, n-1
            while(True):
                mid_a, flag = (left_a + right_a)//2, True
                if(mid_a >= 1):
                    orient = orientation(A[mid_a], B[mid_b], A[mid_a - 1], epsilon)
                    if(orient*p >= 0):
                        right_a, flag = mid_a-1, False
                if(mid_a <= n-2):
                    orient = orientation(A[mid_a], B[mid_b], A[mid_a + 1], epsilon)
                    if(orient*p > 0):
                        left_a, flag = mid_a+1, False
                if(flag):
                    break  
            if(check(A, B, mid_a, mid_b, p, epsilon)):
                return mid_a, mid_b
    
    
    def recursion(points, epsilon):
        n = len(points)
        if (n <= 2):
            return points
        mid, ind_a_up, ind_b_up, ind_a_down, ind_b_down = n//2, -1, -1, -1, -1
        A = recursion(points[0:mid], epsilon)
        B = recursion(points[mid:], epsilon)
        A_up, A_down = split(A)
        B_up, B_down = split(B)    
        ind_a_up, ind_b_up = bin_search(A_up, B_up, 1, epsilon)
        up_total = merge(A_up, B_up, ind_a_up, ind_b_up)
        ind_a_down, ind_b_down = bin_search(A_down, B_down, -1, epsilon)
        down_total = merge(A_down, B_down, ind_a_down, ind_b_down)
        res = concat_arr(up_total, down_total)
        return res
    

    points = copy.deepcopy(Q)
    points.sort()
    convex_hull = recursion(points, epsilon)
    if clockwise == False:
        convex_hull.reverse()
    return convex_hull
    

    '''
    Funkcja buduje otoczkę wypukłą dla podanego
    zbioru punktów Q algorymem Jarvisa
    :parm Q: zbiór punktów
    :return: tablica punktów w postaci krotek współrzędnych
    '''
```

```python
def divide_and_conquer_algorithm_draw_merge(Q,epsilon = 1e-25):
    
    def orientation(a,b,c, epsilon):
        if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
            b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
            c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1
    
    
    def merge(A, B, a, b):
        n, m, hull = len(A), len(B), []
        for i in range(a+1):
            hull.append(A[i])
        for j in range(b, m, 1):
            hull.append(B[j])
        return hull
    
    
    def concat_arr(arr1, arr2):
        res = []
        for i in range(len(arr1)):
            res.append(arr1[i])
        m = len(arr2)
        start, end = 0, m-1
        if(arr2[0] == arr1[0]):
            start = 1
        if(arr2[-1] == arr2[-1]):
            end = m-2
        for j in range(end, start-1, -1):
            res.append(arr2[j])
        return res
    
    
    def split(hull):
        n, ind = len(hull), -1
        for i in range(1, n, 1):
            if(hull[i-1] > hull[i]):
                ind = i-1
                break
        if(ind == -1):
            return hull, [hull[0], hull[-1]]
        l, r = hull[:(ind+1)], hull[(ind):]
        r.append(hull[0])
        r.reverse()
        return l, r
        
    
    def check(A, B, mid_a, mid_b, p, epsilon):
        n, m = len(A), len(B)
        if (mid_a >= 1):
            orient = orientation(A[mid_a], B[mid_b], A[mid_a - 1], epsilon)
            if (orient * p >= 0):
                return False
        if (mid_a <= n-2):
            orient = orientation(A[mid_a], B[mid_b], A[mid_a + 1], epsilon)
            if (orient * p > 0):
                return False
        if (mid_b >= 1):
            orient = orientation(A[mid_a], B[mid_b], B[mid_b - 1], epsilon)
            if (orient * p > 0):
                return False
        if (mid_b <= m-2):
            orient = orientation(A[mid_a], B[mid_b], B[mid_b + 1], epsilon)
            if (orient * p >= 0):
                return False
        return True
        
    
    def bin_search(A, B, p, epsilon):
        n, m = len(A), len(B)
        left_a, right_a, left_b, right_b, mid_a, mid_b = 0, n-1, 0, m-1, -1, -1
        mid_a = (left_a + right_a)//2
        while(True):
            left_b, right_b = 0, m-1
            while(True):
                mid_b, flag = (left_b + right_b)//2, True
                if(mid_b >= 1):
                    orient = orientation(A[mid_a], B[mid_b], B[mid_b-1], epsilon)
                    if(orient * p > 0):
                        right_b, flag = mid_b-1, False
                if(mid_b <= m-2):
                    orient = orientation(A[mid_a], B[mid_b], B[mid_b+1], epsilon)
                    if(orient * p >= 0):
                        left_b, flag = mid_b+1, False
                if(flag):
                    break
            left_a, right_a = 0, n-1
            while(True):
                mid_a, flag = (left_a + right_a)//2, True
                if(mid_a >= 1):
                    orient = orientation(A[mid_a], B[mid_b], A[mid_a - 1], epsilon)
                    if(orient*p >= 0):
                        right_a, flag = mid_a-1, False
                if(mid_a <= n-2):
                    orient = orientation(A[mid_a], B[mid_b], A[mid_a + 1], epsilon)
                    if(orient*p > 0):
                        left_a, flag = mid_a+1, False
                if(flag):
                    break  
            if(check(A, B, mid_a, mid_b, p, epsilon)):
                return mid_a, mid_b
    
    
    def Get_special(first,second):
        special = []
        for i in range (len(first)):
            if first[i] in second:
                special.append(first[i])
        return special
    
    
    def vis_hull(hulls,to_color,colors,tangent=[]):
        vis.clear()
        before = []
        for i in range (len(to_color)):
            vis.add_point([to_color[i][j] for j in range (len(to_color[i])) if to_color[i][j] not in before], color=[colors[i]])
            before.extend(to_color[i])
        for i in range (len(hulls)):
            if len(hulls[i]) == 2:
                vis.add_line_segment((hulls[i][0], hulls[i][1]), color=['blue'])
            elif len(hulls[i]) >= 3:
                vis.add_line_segment([(hulls[i][j], hulls[i][j+1]) for j in range(len(hulls[i])-1)], color=['blue'])
                vis.add_line_segment((hulls[i][0], hulls[i][len(hulls[i])-1]), color=['blue'])
        for i in range (len(tangent)):
            vis.add_line(tangent[i], color = "red")
        vis.show()
    
    
    def recursion(points, epsilon, all_points, rek):
        n = len(points)
        if (n <= 2):
            return points
        mid, ind_a_up, ind_b_up, ind_a_down, ind_b_down = n//2, -1, -1, -1, -1
        A = recursion(points[0:mid], epsilon, all_points, rek+1)
        B = recursion(points[mid:], epsilon, all_points, rek+1)
        hulls = [A,B]
        if rek == 1:
            vis_hull(hulls,[A,B,all_points], ['orange','violet','black'])
        
        A_up, A_down = split(A)
        B_up, B_down = split(B)
        if rek == 1:
            special_A = Get_special(A_up,A_down)
            special_B = Get_special(B_up,B_down)
            special = special_A + special_B
            vis_hull(hulls,[special,A_up,A_down,B_up,B_down,all_points], ['brown','orange','blue','violet','cyan','black'])
        
        ind_a_up, ind_b_up = bin_search(A_up, B_up, 1, epsilon)
        if rek == 1:
            tangent_points = [A_up[ind_a_up], B_up[ind_b_up]]
            tangent = [(A_up[ind_a_up], B_up[ind_b_up])]
            vis_hull(hulls,[tangent_points,special,A_up,B_up,A_down,B_down,all_points], ['grey','brown','orange','violet','blue','cyan','black'], tangent)
        
        up_total = merge(A_up, B_up, ind_a_up, ind_b_up)
        
        ind_a_down, ind_b_down = bin_search(A_down, B_down, -1, epsilon)
        if rek == 1:
            tangent_points.extend([A_down[ind_a_down], B_down[ind_b_down]])
            tangent.append((A_down[ind_a_down], B_down[ind_b_down]))
            vis_hull(hulls,[tangent_points,special,A_up,B_up,A_down,B_down,all_points], ['grey','brown','orange','violet','blue','cyan','black'], tangent)
            
        down_total = merge(A_down, B_down, ind_a_down, ind_b_down)
    
        res = concat_arr(up_total, down_total)
        if rek == 1:
            the_rest = A_down + B_down + A_up + B_up
            vis_hull(hulls,[res,the_rest,all_points], ['green','red','black'], tangent)
            vis_hull([res],[res,all_points], ['blue','black'])
        return res


    points = copy.deepcopy(Q)
    points.sort()
    convex_hull = recursion(points, epsilon, points, 1)


```

```python
def divide_and_conquer_algorithm_draw_binary_search(Q,epsilon = 1e-25):
    
    def orientation(a,b,c, epsilon):
        if epsilon < 1e-10:
            # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
            a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
            b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
            c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
        det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
        if abs(det)<epsilon:
            return 0
        elif det>0:
            return 1
        else:
            return -1
    
    
    def merge(A, B, a, b):
        n, m, hull = len(A), len(B), []
        for i in range(a+1):
            hull.append(A[i])
        for j in range(b, m, 1):
            hull.append(B[j])
        return hull
    
    
    def concat_arr(arr1, arr2):
        res = []
        for i in range(len(arr1)):
            res.append(arr1[i])
        m = len(arr2)
        start, end = 0, m-1
        if(arr2[0] == arr1[0]):
            start = 1
        if(arr2[-1] == arr2[-1]):
            end = m-2
        for j in range(end, start-1, -1):
            res.append(arr2[j])
        return res
    
    
    def split(hull):
        n, ind = len(hull), -1
        for i in range(1, n, 1):
            if(hull[i-1] > hull[i]):
                ind = i-1
                break
        if(ind == -1):
            return hull, [hull[0], hull[-1]]
        l, r = hull[:(ind+1)], hull[(ind):]
        r.append(hull[0])
        r.reverse()
        return l, r
        
    
    def check(A, B, mid_a, mid_b, p, epsilon):
        n, m = len(A), len(B)
        if (mid_a >= 1):
            orient = orientation(A[mid_a], B[mid_b], A[mid_a - 1], epsilon)
            if (orient * p >= 0):
                return False
        if (mid_a <= n-2):
            orient = orientation(A[mid_a], B[mid_b], A[mid_a + 1], epsilon)
            if (orient * p > 0):
                return False
        if (mid_b >= 1):
            orient = orientation(A[mid_a], B[mid_b], B[mid_b - 1], epsilon)
            if (orient * p > 0):
                return False
        if (mid_b <= m-2):
            orient = orientation(A[mid_a], B[mid_b], B[mid_b + 1], epsilon)
            if (orient * p >= 0):
                return False
        return True
        
    
    def vis_hull(hulls,to_color,colors, extra=[], tangents=[], line_color='red'):
        vis.clear()
        before = []
        if len(tangents) > 0:
            vis.add_line(tangents, color = line_color)
        if len(extra) > 0:
            the_line = extra.pop()
            vis.add_line(the_line, color = line_color)
        for i in range (len(hulls)):
            if len(hulls[i]) == 2:
                vis.add_line_segment((hulls[i][0], hulls[i][1]), color=['blue'])
            elif len(hulls[i]) >= 3:
                vis.add_line_segment([(hulls[i][j], hulls[i][j+1]) for j in range(len(hulls[i])-1)], color=['blue'])
                vis.add_line_segment((hulls[i][0], hulls[i][len(hulls[i])-1]), color=['blue'])
        for i in range (len(extra)):
            vis.add_point([extra[i][0]], color=[extra[i][1]])
            before.append(extra[i][0])
        for i in range (len(to_color)):
            vis.add_point([to_color[i][j] for j in range (len(to_color[i])) if to_color[i][j] not in before], color=[colors[i]])
            before.extend(to_color[i])
        vis.show()
    
    
    def bin_search(A, B, p, epsilon, rek, the_rest=[], all_points=[], hulls=[]):
        n, m = len(A), len(B)
        left_a, right_a, left_b, right_b, mid_a, mid_b = 0, n-1, 0, m-1, -1, -1
        mid_a = (left_a + right_a)//2
        while(True):
            left_b, right_b = 0, m-1
            while(True):
                mid_b, flag = (left_b + right_b)//2, True
                if rek == 1:
                    borders = [B[left_b], B[right_b]]
                    colors = ['brown','cyan','grey','orange','violet','blue','black']
                    vis_hull(hulls,[[B[mid_b]],[A[mid_a]],borders,A,B,the_rest,all_points], colors)
                    extra = []
                    extra_check = []
                if(mid_b >= 1):
                    flag_vis = True
                    orient = orientation(A[mid_a], B[mid_b], B[mid_b-1], epsilon)
                    if(orient * p > 0):
                        right_b, flag = mid_b-1, False
                        flag_vis = False
                    if rek == 1:
                        if flag_vis:
                            extra.append([B[mid_b-1], 'green'])
                        else:
                            extra.append([B[mid_b-1], 'red'])
                if(mid_b <= m-2):
                    flag_vis = True
                    orient = orientation(A[mid_a], B[mid_b], B[mid_b+1], epsilon)
                    if(orient * p >= 0):
                        left_b, flag = mid_b+1, False
                        flag_vis = False
                    if rek == 1:
                        if flag_vis:
                            extra.append([B[mid_b+1], 'green'])
                        else:
                            extra.append([B[mid_b+1], 'red'])
                if rek == 1:
                    extra.append((A[mid_a],B[mid_b]))
                    extra_copy = copy.deepcopy(extra)
                    colors = ['brown','cyan','grey','orange','violet','blue','black']
                    line_color = 'green'
                    if extra[0][1] == 'red' or extra[1][1] == 'red':
                        line_color = 'red'
                    vis_hull(hulls,[[B[mid_b]],[A[mid_a]],borders,A,B,the_rest,all_points], colors, extra_copy, line_color=line_color)
                if(flag):
                    if rek == 1:
                        extra_check = copy.deepcopy(extra)
                        extra_check.pop()
                    break
            left_a, right_a = 0, n-1
            while(True):
                mid_a, flag = (left_a + right_a)//2, True
                if rek == 1:
                    borders = [A[left_a], A[right_a]]
                    colors = ['brown','cyan','grey','orange','violet','blue','black']
                    vis_hull(hulls,[[A[mid_a]],[B[mid_b]],borders,A,B,the_rest,all_points], colors)
                    extra=[]
                if(mid_a >= 1):
                    flag_vis = True
                    orient = orientation(A[mid_a], B[mid_b], A[mid_a - 1], epsilon)
                    if(orient*p >= 0):
                        right_a, flag = mid_a-1, False
                        flag_vis = False
                    if rek == 1:
                        if flag_vis:
                            extra.append([A[mid_a-1], 'green'])
                        else:
                            extra.append([A[mid_a-1], 'red'])
                if(mid_a <= n-2):
                    flag_vis = True
                    orient = orientation(A[mid_a], B[mid_b], A[mid_a + 1], epsilon)
                    if(orient*p > 0):
                        left_a, flag = mid_a+1, False
                        flag_vis = False
                    if rek == 1:
                        if flag_vis:
                            extra.append([A[mid_a+1], 'green'])
                        else:
                            extra.append([A[mid_a+1], 'red'])
                if rek == 1:
                    extra.append((A[mid_a],B[mid_b]))
                    extra_copy = copy.deepcopy(extra)
                    colors = ['brown','cyan','grey','orange','violet','blue','black']
                    line_color = 'green'
                    if extra[0][1] == 'red' or extra[1][1] == 'red':
                        line_color = 'red'
                    vis_hull(hulls,[[A[mid_a]],[B[mid_b]],borders,A,B,the_rest,all_points], colors, extra_copy, line_color=line_color)
                if(flag):
                    if rek == 1:
                        extra_check.extend(copy.deepcopy(extra))
                    break  
            if(check(A, B, mid_a, mid_b, p, epsilon)):
                if rek == 1:
                    colors = ['brown','cyan','orange','violet','blue','black']
                    vis_hull(hulls,[[A[mid_a]],[B[mid_b]],A,B,the_rest,all_points], colors, extra_check, line_color='green')            
                return mid_a, mid_b
    
    
    def recursion(points, epsilon, all_points, rek):
        n = len(points)
        if (n <= 2):
            return points
        mid, ind_a_up, ind_b_up, ind_a_down, ind_b_down = n//2, -1, -1, -1, -1
        A = recursion(points[0:mid], epsilon, all_points, rek+1)
        B = recursion(points[mid:], epsilon, all_points, rek+1)
        if rek == 1:
            hulls = [A,B]
            vis_hull(hulls,[A,B,all_points],['orange','violet','black'])
        
        A_up, A_down = split(A)
        B_up, B_down = split(B)
        if rek == 1:
            the_rest = A_down + B_down
            ind_a_up, ind_b_up = bin_search(A_up, B_up, 1, epsilon, rek, the_rest, all_points, hulls)
        else:
            ind_a_up, ind_b_up = bin_search(A_up, B_up, 1, epsilon, rek)
        the_rest = A_up + B_up
        up_total = merge(A_up, B_up, ind_a_up, ind_b_up)
        
        if rek == 1:
            the_rest = A_up + B_up
            ind_a_down, ind_b_down = bin_search(A_down, B_down, -1, epsilon, rek, the_rest, all_points, hulls)
            all_on_hull = A_up + B_up + A_down + B_down
            tangentPoints = [A_up[ind_a_up],B_up[ind_b_up],A_down[ind_a_down],B_down[ind_b_down]]
            tangents = [(A_up[ind_a_up],B_up[ind_b_up]), (A_down[ind_a_down],B_down[ind_b_down])]
            vis_hull(hulls,[tangentPoints,all_on_hull,all_points], ['green','blue','black'], tangents = tangents, line_color='green')
        else:
            ind_a_down, ind_b_down = bin_search(A_down, B_down, -1, epsilon, rek)
        the_rest = the_rest + A_down + B_down
        down_total = merge(A_down, B_down, ind_a_down, ind_b_down)
        
        res = concat_arr(up_total, down_total)
        return res


    points = copy.deepcopy(Q)
    points.sort()
    convex_hull = recursion(points, epsilon, points, 1)


```

```python
def Start_point (Q):
    start=Q[0]
    start_idx=0
    for i in range (1,len(Q)):
        if Q[i][1]<start[1]:
            start=Q[i]
            start_idx=i
        elif Q[i][1]==start[1]:
            if Q[i][0]<start[0]:
                start=Q[i]
                start_idx=i
    return start,start_idx


def orientation (a,b,c,epsilon):
    if epsilon < 1e-10:
        # Występuje tu zamiana typu danych na dokładność do 64 liczb miejsc po przecinku
        a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
        b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
        c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
    det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
    if abs(det)<epsilon:
        return 0
    elif det>0:
        return 1
    else:
        return -1


def further (a,b,c):
    if a[0]==b[0]:
        if abs(a[1]-b[1])>abs(a[1]-c[1]):
            return True
        else:
            return False
    else:
        if abs(a[0]-b[0])>abs(a[0]-c[0]):
            return True
        else:
            return False


def ComparePoints_v2(start_point, first, second, epsilon):
    orient=orientation(start_point, first, second, epsilon)
    if orient>0 or (orient==0 and further(start_point, first, second)==False):
        return -1
    else:
        return 1
    

def Sort_points (Q,start,start_idx,epsilon):
    if len(Q)==1: return Q
    start_point = Q[start_idx]
    Q.pop(start_idx)    
    Q = sorted(Q, key = functools.cmp_to_key(lambda first, second: ComparePoints_v2(start_point, first, second, epsilon)))
    Q_sorted=[start_point]
    for i in range (len(Q)-1):
        if orientation(start_point,Q[i],Q[i+1], epsilon)!=0:
            Q_sorted.append(Q[i])
    Q_sorted.append(Q[len(Q)-1])
    return Q_sorted


def Get_convex_hull(Q,epsilon):
    convex_hull=[Q[0],Q[1],Q[2]]
    i=3
    t=2
    while i<len(Q):
        if orientation(convex_hull[t-1],convex_hull[t],Q[i],epsilon)>0:
            convex_hull.append(Q[i])
            t+=1
            i+=1
        else:
            convex_hull.pop()
            t-=1
    return convex_hull



def graham_algorithm(Q, epsilon = 1e-25, clockwise = False):
    points = copy.deepcopy(Q)
    start,start_idx=Start_point(points)
    points=Sort_points(points,start,start_idx,epsilon)
    if len(points)<3:
        convex_hull = points
    else:
        convex_hull = Get_convex_hull(points,epsilon)
    if clockwise == True:
        convex_hull.reverse()
    return convex_hull


def add_perimeter(points, vis, col = "red"):
    n = len(points)
    for i in range(n-1):
        vis.add_line_segment([(points[i], points[i+1])], color = col)
    vis.add_line_segment([(points[-1], points[0])], color = col)
    
    
def further (a,b,c):
    if a[0]==b[0]:
        if abs(a[1]-b[1])>abs(a[1]-c[1]):
            return True
        else:
            return False
    else:
        if abs(a[0]-b[0])>abs(a[0]-c[0]):
            return True
        else:
            return False

    
def Orientation_v2(p1, p2, p3, epsilon = 0):
        x1, y1, x2, y2, x3, y3 = *p1, *p2, *p3
        d = (y3-y2)*(x2-x1) - (y2-y1)*(x3-x2)
        if d > 0:
            return 1
        elif d < 0:
            return -1
        else:
            return 0
        
def ComparePoints(p1, first, second, epsilon = 0):
    orient= Orientation_v2(p1, first, second, epsilon)
    if orient>0 or (orient==0 and further(p1, first, second)== True):
        # first is more to the right
        return 1
    else:
        return -1
        
        
def calc_orient(convex, p1,  x, n):
        return ComparePoints(p1, convex[x], convex[(x+1) % n]), ComparePoints(p1, convex[x], convex[(x-1) % n])

    
def draw_point(p1, convex, mid, vis):
    vis.add_point([convex[mid]], color = "lime")
    vis.add_line([(p1, convex[mid])], color = "lime")
    vis.show()
    return


def bin_search(p1, convex, vis, show = True):
        
        n = len(convex)
        l, r = 0, n//2
        INF = float('inf')
        count = 0
        #print (convex)
        
            
        while(True):
            l1, l2 = calc_orient(convex, p1, l, n) #odwracamy zeby latwiej sie liczylo - tak jakby katy byly liczbami
            r1, r2 = calc_orient(convex, p1, r, n)  #to samo
            l1, l2, r1, r2 = l1 * (-1), l2 * (-1), r1 * (-1), r2 * (-1)
            
            #print("l:", l)
            #print("r:", r)
            
            
            if(show):
                vis.add_point(convex, color = "blue")
                #add_perimeter(convex, vis, "blue")
                vis.add_point([p1], color = "red")
                vis.add_point([convex[l]], color = "magenta")
                vis.add_point([convex[r]], color = "cyan")
                vis.show()
            
            if(l1 < 0 and l2 < 0):
                return l, INF

            if(r1 < 0 and r2 < 0):
                return r, INF

            #Przy kazdym z osobna wyznaczaniu mid trzeba zweryfikowac, czy nie jest on moze szukana styczna

            if(l1 < 0 and l2 > 0):

                if(r1 < 0 and r2 > 0):

                    #sprawdzac czy left czy right jest wczesniej w lancuchu malejacym i na tej podstawie wybierac czesc

                    orient_l_r = ComparePoints(p1, convex[l], convex[r])

                    #orient_l_r < 0 oznacza, że kąt dla r jest wiekszy

                    if(orient_l_r < 0): #jesli right jest "wyzej" wtedy zamieniamy kolejnosc, bo chcemy zeby: kat dla r < kat dla l
                        temp, l = l, r
                        r = temp

                    mid = (r + ((l-r) % n)//2 ) % n 
                    if(show):
                        draw_point(p1, convex, mid, vis)

                    #sprawdzamy czy mid nalezy do tego samego lancucha co l i r

                    mid1, mid2 = calc_orient(convex, p1, mid, n)
                    mid1, mid2 = mid1 * (-1), mid2 * (-1)

                    if(mid1 < 0 and mid2 < 0):
                        return mid, INF

                    if(mid1 < 0 and mid2 > 0):
                        orient_r_mid = ComparePoints(p1, convex[r], convex[mid])

                        if(orient_r_mid < 0):
                            l, r = r, mid

                        else:
                            l, r = mid, l

                    else:
                        l, r = mid, l

                elif(r1 > 0 and r2 < 0):

                    #idziemy w strone plusow,  ale potem dla mid trzeba sprawdzic
                    #w ktorym lancuchu sie znajdzie i na tej podstawie okreslac czy isc na (right, mid) czy (mid, left)
                    
                    mid = (r + ((l-r)% n)//2 ) % n
                    
                    if(show):
                        draw_point(p1, convex, mid, vis)

                    mid1, mid2 = calc_orient(convex, p1, mid, n)
                    mid1, mid2 = mid1 * (-1), mid2 * (-1)

                    if(mid1 < 0 and mid2 < 0):
                        return mid, INF

                    if(mid1 < 0 and mid2 > 0):
                        l, r = r, mid

                    else:
                        l, r = mid, l

                elif(r1 > 0 and r2 > 0):

                    #idziemy w strone plusow
                    #musimy sprawdzic w ktorym lancuchu  znajdzie sie mid
                    
                    mid = (r + ((l-r) % n)//2 ) % n
                    
                    if(show):
                        draw_point(p1, convex, mid, vis)

                    mid1, mid2 = calc_orient(convex, p1, mid, n)
                    mid1, mid2 = mid1 * (-1), mid2 * (-1)

                    if(mid1 < 0 and mid2 < 0):
                        return mid, INF

                    if(mid1 < 0 and mid2 > 0):
                        l, r = r, mid

                    elif(mid1 > 0 and mid2 < 0):
                        l, r = mid, l
                        


            elif(l1 > 0 and l2 < 0):

                if(r1 < 0 and r2 > 0):
                    #praktycznie to samo co w poprzednim ifie, tylko zamienione miejscami
                    temp, l = l, r
                    r = temp
                    mid = (r + ((l-r) % n)//2 ) % n
                    
                    if(show):
                        draw_point(p1, convex, mid, vis)

                    mid1, mid2 = calc_orient(convex, p1, mid, n)
                    mid1, mid2 = mid1 * (-1), mid2 * (-1)

                    if(mid1 < 0 and mid2 < 0):
                        return mid, INF

                    if(mid1 > 0 and mid2 < 0):
                        l, r = mid, l

                    else:
                        l, r = r, mid


                elif(r1 > 0 and r2 < 0):

                    #sprawdzic czy left czy right jest wczesniej w lancuchu rosnacym, przypadek z poprzedniego ifa
                    #tylko zamieniony miejscami
                    orient_l_r = ComparePoints(p1, convex[l], convex[r])

                    if(orient_l_r > 0): #right jest pozniej w lancuchu rosnacym wiec musi wiec wiekszy zatem, zatem orient < 0
                        temp, l = l, r
                        r = temp

                    mid = (r + ((l-r) % n)//2 ) % n
                    
                    if(show):
                        draw_point(p1, convex, mid, vis)

                    mid1, mid2 = calc_orient(convex, p1, mid, n)
                    mid1, mid2 = mid1 * (-1), mid2 * (-1)

                    if(mid1 < 0 and mid2 < 0):
                        return mid, INF

                    if(mid1 > 0 and mid2 < 0): #jesli mid tez jest w lancuchu rosnacym wtedy sprawdzamy jego orientacje wzgledem r

                        orient_r_mid = ComparePoints(p1, convex[r], convex[mid])

                        if(orient_r_mid < 0): #to oznacza ze punkt jest powyzej right w lancuchu rosnacym, zatem bierzemy mid, l
                            l, r = mid, l

                        else: #Jesli jest ponizej lefta w lancuchu rosnacym, wtedy bierzemy b
                            l, r = r, mid

                    elif(mid1 < 0 and mid2 > 0): #jesli l i r sa w lancuchu roznacym a mid w malejacym to bierzemy r, mid
                        l, r = r, mid

                    elif(mid1 > 0 and mid2 > 0): #tak samo gdy mid jest "dolną" styczną             
                        l, r = r, mid


                elif(r1 > 0 and r2 > 0):
                    #korzystam z rozwiazania na dole, tylko zamieniam zmienne

                    temp, l = l, r
                    r = temp

                    mid = (r + ((l-r) % n)//2 ) % n
                    
                    if(show):
                        draw_point(p1, convex, mid, vis)

                    mid1, mid2 = calc_orient(convex, p1, mid, n)
                    mid1, mid2 = mid1 * (-1), mid2 * (-1)

                    if(mid1 < 0 and mid2 < 0):
                        return mid, INF

                    if(mid1 > 0 and mid2 < 0):
                        l, r = mid, l

                    else:
                        l, r = r, mid
            
            elif(l1 > 0 and l2 > 0):

                if(r1 < 0 and r2 > 0):
                    #rozpatrzone wcześniej, wystarczy zammienić left i right miejscami
                    temp, l = l, r
                    r = temp
                    mid = (r + ((l-r) % n)//2 ) % n   

                    if(show):
                        draw_point(p1, convex, mid, vis)

                    mid1, mid2 = calc_orient(convex, p1, mid, n)
                    mid1, mid2 = mid1 * (-1), mid2 * (-1)

                    if(mid1 < 0 and mid2 < 0):
                        return mid, INF

                    if(mid1 > 0 and mid2 < 0):
                        l, r = mid, l

                    elif(mid1 < 0 and mid2 > 0):
                        l, r = r, mid

                elif(r1 > 0 and r2 < 0):
                    #skorzystac z tego tutaj wyzej przy obliczaniu analogicznej sytuacji tylko z zamenionym l i r
                    mid = (r + ((l-r) % n)//2 ) % n
                    
                    if(show):
                        draw_point(p1, convex, mid, vis)

                    mid1, mid2 = calc_orient(convex, p1, mid, n)
                    mid1, mid2 = mid1 * (-1), mid2 * (-1)


                    if(mid1 < 0 and mid2 < 0):
                        return mid, INF

                    if(mid1 > 0 and mid2 < 0):
                        l, r = mid, l

                    elif(mid1 < 0 and mid2 > 0):
                        l, r = r, mid
                        
            if(show):
                vis.clear()
                vis.add_point(convex, color = "blue")
                #add_perimeter(convex, vis, "blue")
                vis.add_point([p1], color = "red")
                vis.add_point([convex[l], convex[r]], color = "orange")  
                vis.show()
                vis.clear()
                
            count += 1
```

```python
global colors
colors = ["red", "green", "orange", "brown", "grey", "pink", "black", "blue", "lime", "cyan", "gold", "magenta"]


def Chan_algorithm(Q, show = False):
    #print (Q)
    n = len(Q)
    t = 4
    
    
    def dist(p1, p2):
        x1, y1, x2, y2 = *p1, *p2
        return (y2-y1)**2 + (x2-x1)**2
    
    def orientation(p1, p2, p3):
        x1, y1, x2, y2, x3, y3 = *p1, *p2, *p3
        d = (y3-y2)*(x2-x1) - (y2-y1)*(x3-x2)
        if d > 0:
            return 1
        elif d < 0:
            return -1
        else:
            return 0
    
    
    def check_mid(convex, p1, mid, n):
        x1, x2 = calc_orient(convex, p1, mid, n)
        if(x1 < 0 and x2 < 0):
            return True
        return False
    
    
        
    def max_angle(p2, p1, points):
        test, n = points[0], len(points)
        for i in range(1, n, 1):
            orient = orientation(p1, test, points[i])
            if(orient < 0 or ( orient == 0 and dist(p1, test) < dist(p1, points[i]) ) ):
                test = points[i]
        return test
    
    
    def min_point(p1, p2):
        if(p1[1] < p2[1] or (p1[1] == p2[1] and p1[0] <= p2[0] )):
            return p1
        return p2
        
        
    def find_bottom(Q):
        INF = float('inf')
        min_p = (INF, INF)
        n = len(Q)
        for i in range(n):
            min_p = min_point(min_p, Q[i])
        return min_p
    
    
    def chan(points, m):
        global colors
        vis2, vis3 = Visualizer(), Visualizer()
        k, start = n//m + (n % m != 0), 0
        angles, col = [0] * k, 0
        INF = -float('inf')
        P, convex = [], [(INF, 0)]
        convex.append(find_bottom(points))
        
        if(show):
            vis2.add_point(points, color = "blue")
            vis2.show()
        
        for i in range(k):
            P.append(Q[start:(min(start+m, n))])
            start += m
            
            if(show):
                vis2.add_point(P[i], color = colors[i%12])

        if(show):  
            vis2.show()
        
        for i in range(k):
            P[i] = graham_algorithm(P[i]) 
            
            if(show):
                add_perimeter(P[i], vis2, colors[i%12])
        
        if(show):
            vis2.show()
        
        for i in range(m):
            for j in range(k):
                if(len(P[j]) <= 2):
                    angles[j] = max_angle(convex[-2], convex[-1], P[j])
                else:
                    vis6 = Visualizer()
                    x, y = bin_search(convex[-1], P[j], vis6, False)
                    angles[j] = P[j][x]
                    
            if(show):
                vis2.clear()
                vis2.add_point(points, color = "blue")
                vis2.add_point(angles, color = "black")
                vis2.add_point(convex[1:], color = "orange")
                vis2.show()
                for jj in range(k):
                    add_perimeter(P[jj], vis2, "blue")
                vis2.add_point(angles, color = "black")
                vis2.add_point(convex[1:], color = "orange")
                vis2.show()
        
            
            max_point = max_angle(convex[-2], convex[-1], angles)
            if(max_point == convex[1]):
                if (show):
                    print("ZNALEZIONE")
                    vis3.add_point(points, color = "blue")
                    vis3.add_point(convex[1:], color = "red")
                    add_perimeter(convex[1:], vis3, "red")
                    vis3.show()
                return True, convex[1:]
            convex.append(max_point)
            
        return False, []
    
    
    def run(Q):
        t, n = 2, len(Q)
        if(n < 3):
            return Q
        t1 = time.perf_counter()
        while(True):
            t = min(n, t)
            #print(t)
            x, y = chan(Q, t)
            if(x):
                t2 = time.perf_counter()
                #print("Czas Wykonania:", t2-t1)
                return y
            if(t >= n):
                break
            t *= t
            
    wyn = run(Q)    
    return wyn
```

# Porównanie czasu działania algorytmów

Porównnie czasu działania funkcji wyznaczającej orienację punktu względem prostej na punktach typu float z punktami typu Decimal posiadająe precyzją do 32 miejsc po przecinku

```python
def orient_decimal (a,b,c):
    eps=eps=1e-25
    new_a=(decimal.Decimal(a[0]),decimal.Decimal(a[1]))
    new_b=(decimal.Decimal(b[0]),decimal.Decimal(b[1]))
    new_c=(decimal.Decimal(c[0]),decimal.Decimal(c[1]))
    det=(new_b[0]-new_a[0])*(new_c[1]-new_b[1])-(new_b[1]-new_a[1])*(new_c[0]-new_b[0])
    if abs(det)<eps:
        return 0
    elif det>0:
        return 1
    else:
        return -1

def orient_float (a,b,c):
    eps=eps=1e-8
    det=(b[0]-a[0])*(c[1]-b[1])-(b[1]-a[1])*(c[0]-b[0])
    if abs(det)<eps:
        return 0
    elif det>0:
        return 1
    else:
        return -1


a=(1,1)
b=(1,2)
c=(2,3)
time_decimal=0
time_float=0
for i in range (int(1e6)):
    start_time_decimal = time.time()
    _ = orient_decimal(a,b,c)
    end_time_decimal = time.time()
    time_decimal += end_time_decimal - start_time_decimal
    start_time_float = time.time()
    _ = orient_float(a,b,c)
    end_time_float = time.time()
    time_float += end_time_float - start_time_float

table_orient = pd.DataFrame([time_decimal,time_float], index=["Decimal","Float"], columns=["Czas"])
display (table_orient)
```

```python
def Generate_time(points_function, k, file_path = None):

    def GenerateResult(points, function):
        start_time = time.time()
        _ = function(points)
        end_time = time.time()
        time_result = round(end_time - start_time, 3)
        return time_result
    
    
    time_data=[[0 for i in range (10)] for _ in range (7)] #10
    time_name=[]
    for i in range (10):
        n=(i+1)*k
        points = points_function(n=n)
        time_name.append(str(n)+" Punktów")
        time_data[0][i] = GenerateResult(points, graham_algorithm)
        time_data[1][i] = GenerateResult(points, jarvis_algorithm)
        time_data[2][i] = GenerateResult(points, incremental_algorithm)
        time_data[3][i] = GenerateResult(points, andrews_monotone_chain_algorithm)
        time_data[4][i] = GenerateResult(points, quickhull_algorithm)
        time_data[5][i] = GenerateResult(points, divide_and_conquer_algorithm)
        time_data[6][i] = GenerateResult(points, Chan_algorithm)
    
    index_names = ["Graham","Jarvis","Przyrostowa","Górna i dolna otoczka","QuickHull","Dziel i zwyciężaj", "Chan_algorithm"]
    table_time = pd.DataFrame(time_data, index=index_names, columns=time_name)
    display (table_time)
    if file_path != None:
        table_time.to_csv(file_path)
```
